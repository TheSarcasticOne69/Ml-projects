**10. write a program to implement the decision tree classification algorithm for a sample training data set stored as a .csv file. Conclude the prediction**

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics

# Step 1: Load the dataset
data = pd.read_csv('data.csv')

# Step 2: Encode the target column ('Purchased') to numeric
data['Purchased'] = data['Purchased'].map({'No': 0, 'Yes': 1})

# Step 3: Split dataset into features and target variable
X = data[['Age', 'Salary']]
y = data['Purchased']

# Step 4: Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)

# Step 5: Create Decision Tree classifier object
clf = DecisionTreeClassifier()

# Step 6: Train Decision Tree Classifier
clf = clf.fit(X_train, y_train)

# Step 7: Predict the response for test dataset
y_pred = clf.predict(X_test)

# Step 8: Model Accuracy
accuracy = metrics.accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Step 9: Make a new prediction (e.g., Age: 40, Salary: 45000)
sample_input = [[40, 45000]]
prediction = clf.predict(sample_input)
print("Prediction for sample input:", "Yes" if prediction[0] == 1 else "No")
```
